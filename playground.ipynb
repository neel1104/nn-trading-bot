```python
import yfinance as yf
import pandas as pd

# 1. Fetch Data
ticker = "BTC-USD"
df = yf.download(ticker, period="1y", interval="1wk")
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.droplevel(1)

# 2. Run Functional Pipeline
analyzer = (
    MarketStructureAnalyzer(df)
    .filter_inside_bars()
    .identify_pivots(min_candles=2) # default is set as 2.
    .find_bos()
)

# 3. Visualize
# Zoomed into the first 30 days as requested
analyzer.plot(title=f"{ticker} Market Structure (2-2-2 Rule)", zoom_days=52) # zoom_days default value is 30
```
